```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, f1_score
```

```python
feature_file = "secret/luhft/2023-03-27/AI data with features 3 seconds.csv"

csv = pd.read_csv(feature_file)

labels = np.array(csv["Classification"])
positive_case = "C"
text_label_to_binary = lambda label: 1 if label == positive_case else 0
numeric_labels = np.array([text_label_to_binary(label) for label in labels]).astype("int")
numeric_labels
```

```python
from sklearn.utils import class_weight
cw = class_weight.compute_class_weight( class_weight='balanced', classes=np.unique(numeric_labels), y=numeric_labels)
cw_dict = dict(enumerate(cw))
print(cw_dict)
```

```python
csv = csv.dropna(axis=1)
features = csv.drop(['Study Number', 'Classification', 'Start ah', 'filepath'], axis=1)
features
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score

def run_trial(random_state: int):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

    results = []
    y_true_all = []
    y_pred_all = []
    fold_importance = []

    target = pd.Series(numeric_labels)

    for i, (train_index, test_index) in enumerate(skf.split(features, target)):
        # print(f"Step {i}")
        #print(f"-> test index: {test_index}, train index: {train_index}")
        #print(numeric_labels[test_index])

        X_train, X_test = features.iloc[train_index], features.iloc[test_index]
        y_train, y_test = target.iloc[train_index], target.iloc[test_index]

        rf = RandomForestClassifier(n_estimators=100, max_depth=5, bootstrap=False, random_state=random_state, min_samples_leaf=10, class_weight=cw_dict)

        rf.fit(X_train, y_train)

        y_pred = rf.predict(X_test)
        results.append((y_test.values, y_pred))
        y_true_all.extend(y_test.values)
        y_pred_all.extend(y_pred)

        importance = list(zip(list(features), rf.feature_importances_))
        fold_importance.append(importance)

    cm = confusion_matrix(y_true_all, y_pred_all, labels=[0, 1])
    f1 = f1_score(y_true_all, y_pred_all, average='binary')

    return results, y_true_all, y_pred_all, f1, cm, fold_importance

```

```python
all_fold_importance = []
all_trial_f1 = []
all_tpr = [] 
all_tnr = [] 

for trial in range(0, 10):
    results, y_true_all, y_pred_all, f1, cm, trial_importance = run_trial(random_state=trial)
    print(f"Trial {trial}")
    # print(f"F1 score = {f1}")
    print(cm)

    TN, FP, FN, TP = cm.ravel()
    TPR = TP / (TP + FN)
    all_tpr.append(TPR)
    TNR = TN / (TN + FP)
    all_tnr.append(TNR)

    print(f"True Positive Rate (TPR): {TPR}")
    #print(f"True Negative Rate (TNR): {TNR}")
    #print("")

    all_fold_importance.append(trial_importance)
    all_trial_f1.append(f1)

flattened_pairs_array = [item for sublist in all_fold_importance for item in sublist]

print(f"Mean F1: {np.mean(all_trial_f1)} (min {np.min(all_trial_f1)}, max {np.max(all_trial_f1)})")
print(f"Mean TPR: {np.mean(all_tpr)} (min {np.min(all_tpr)}, max {np.max(all_tpr)})")
print(f"Mean TNR: {np.mean(all_tnr)} (min {np.min(all_tnr)}, max {np.max(all_tnr)})")
```

```python
def importance_summary(importance_scores):
    attributes = {}

    for trial in importance_scores:
        for attr, importance in trial:
            if attr not in attributes:
                attributes[attr] = [importance]
            else:
                attributes[attr].append(importance)

    num_trials = len(importance_scores)

    mean_importance = {attr: np.mean(importances) for attr, importances in attributes.items()}

    # Standard error of the mean
    sem_importance = {attr: np.std(importances, ddof=1) / np.sqrt(num_trials) for attr, importances in attributes.items()}
    
    df = pd.DataFrame({"attribute": list(mean_importance.keys()), 
                       "mean_importance": list(mean_importance.values()),
                       "sem": list(sem_importance.values())})
    
    return df.sort_values(by="mean_importance", ascending=False)


def plot_importance_summary(importance_df):
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.bar(x=importance_df["attribute"], height=importance_df["mean_importance"], yerr=importance_df["sem"], capsize=5)
    ax.set_xticklabels(importance_df["attribute"], rotation=90, fontsize=12)
    ax.set_ylabel("Mean Importance", fontsize=14)
    ax.set_xlabel("Attribute", fontsize=14)
    ax.set_title("Importance Summary", fontsize=16)

    plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

def plot_roc(rf):
    # https://scikit-learn.org/stable/modules/model_evaluation.html#roc-metrics

    y_score = rf.predict_proba(features) # Array [[0-prob 1-prob] ...]
    y_score = y_score[:, 1] # ... but we just want the prob of 1
    fpr, tpr, thresholds = roc_curve(numeric_labels, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange',lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()

plot_roc(rfall)
```
